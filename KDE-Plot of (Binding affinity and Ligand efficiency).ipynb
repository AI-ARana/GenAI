```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts


pd.set_option('display.max_rows',None)
```

```python
mart = pd.read_csv(r"D:\Data_h\Docking Results.csv")

mart.head(10)
```

```python
mart1=mart.drop(['Ligand'],axis=1)
mart1
```

```python
from matplotlib import font_manager
sns.set()

sns.set(style='ticks')

hm=sns.kdeplot(data = mart1,levels=100,  fill=True ,thresh=0.1)

plt.xticks(fontname="Times New Roman",fontsize=12,rotation=0,weight='bold')   
plt.yticks(fontname="Times New Roman",fontsize=12,rotation=0,weight='bold')

plt.xlabel('Binding affinity ΔG(kcal/mol)', fontsize=12, fontname="Times New Roman", weight='bold')
plt.ylabel('Ligand efficiency (LE)(kcal/mol)', fontsize=12,fontname="Times New Roman",weight='bold')


hm=hm.get_figure()
hm.savefig('f:\KDE-Plot of Binding affinity and Ligand efficiency.png', bbox_inches='tight', dpi=300)
```
