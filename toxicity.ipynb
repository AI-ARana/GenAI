```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t = pd.read_csv(r"D:\Data_h\BENZI Toxicity.csv")

t.head(10)
```

```python
t1=t.drop(['Ligand'], axis=1)
t1
```

```python
# Define the plot
fig, ax = plt.subplots(figsize=(16,9))

# Add title to the Heat map
title = "Heat Map"

# Set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl = ax.title
ttl.set_position([0.5,1.05])

cbar_kws = {"orientation":"vertical", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.2, 
           
            "drawedges":True,
           }
# Hide ticks for X & Y axis
#ax.set_xticks([])
#ax.set_yticks([])

# Remove the axes
#ax.axis('off')
# Use the heatmap function from the seaborn package
sns.heatmap(t1,annot= True,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax, square=True, cbar=True, cbar_kws=cbar_kws)

# Display the Pharma Heatmap
plt.show()
```
