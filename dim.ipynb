```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\R&D\dim\data.csv")
print(df.head(5))
```

```python
df.drop('Models', axis=1, inplace=True)
```

```python
# Plot the correlation heatmap
sns.heatmap(df.corr(), cmap="Accent_r", annot=True);
```

```python
fig, ax = plt.subplots(figsize=(13,7))

# Add title to the Heat map
title = "Pharma Sector Heat Map"

# Set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl = ax.title
ttl.set_position([0.5,1.05])

# Hide ticks for X & Y axis
ax.set_xticks([])
ax.set_yticks([])

# Remove the axes
ax.axis('off')
# Use the heatmap function from the seaborn package
sns.heatmap(df,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax)

# Display the Pharma Sector Heatmap
plt.show()
```
