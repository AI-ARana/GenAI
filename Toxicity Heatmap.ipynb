```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
tox1=pd.read_csv('D:\Data_h\BENZI Toxicity.csv')

tox1
```

```python
tox11=tox1.drop(['Ligand'], axis=1)
```
