```python
from matplotlib import pyplot as plt
```

```python
years=[1950,1960,1970,1980,1990,2000,2010]
gdp=[300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3]
```

```python
# Create a line chart , years on x-axis, gdp on y-axisis
plt.plot(years,gdp, color='green',marker='o',linestyle='solid')
# Add Title
plt.title("Nominal GDP")
# add the label to y-axsis
plt.ylabel(" BIllions of $")
plt.show()
```

```python
movies=["Annie Hall", "Ben-Hur","Casablanca","Gandhi","West Side story"]
num_oscars=[5,11,3,8,10]
```

```python
# bar are by default width 0.8, so we'll add 0.1 to left coordinates
# so that each bar is centered
xs=[i+0.5 for i, _ in enumerate(movies)]
plt.bar(xs, num_oscars)

plt.ylabel("# of Academy Awards")
plt.title("My favorite Movies")

# label x_axsis with movie names at bar centre
plt.xticks([i+0.5 for i, _ in enumerate(movies)],movies)

plt.show()
```

```python
grades=[83,95,91,87,70,0,85,82,100,67,73,77,0]
decile=lambda grade:grade//10*10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()],histogram.values(),8)

plt.axis([-5,105,0,5])

plt.sticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam Grades")
plt.show()
```

```python
mentions=[500,505]
years=[2013,2014]

plt.bar([2012.6,2013.6], mentions,0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'Data Science'")

plt.ticklabel_format(useOffset=False)
plt.axis([2012.5,2014.5,499,506])
plt.title("Look at the 'Huge'Increase!")
plt.show()
```

```python
var=[1,2,4,8,16,32,64,128,256]
b_sq=[256,128,64,32,16,8,4,2,1]
total_e=[x+y for x, y in zip(var,b_sq)]
```

```python
var=[1,2,4,8,16,32,64,128,256]
b_sq=[256,128,64,32,16,8,4,2,1]
total_e=[x+y for x, y in zip(var,b_sq)]
xs=[i for i, _ in enumerate(var)]
plt.plot(xs,var, 'g-', label='Variance')
plt.plot(xs,b_sq, 'r-.', label='Bias_Square')
plt.plot(xs, total_e, 'b:', label="Total_ Error")

plt.legend(loc=9)
plt.xlabel("Model Complexity")
plt.title("The Bias-Variance Trdaeoff")
plt.show()
```

```python
friends=[70,65,72,63,71,64,60,64,67]
minu=[175,170,205,120,220,130,105,145,190]
labels=['a','b','c','d','e','f','g','h','i']

plt.scatter(friends,minu)

for label, friend_count, minu_count in zip(labels, friends, minu):
    plt.annotate(label,xy=(friend_count, minu_count),xytext=(5, -5),textcoords='offset points')
plt.title("Daily Minutes vs Number of Friends")
plt.xlabel("# of Friends")
plt.ylabel("Daily minutes spent on the site")
plt.show()
```
