```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("D:\Data_h\Table NRU.csv")
data.describe() 
data.head()
```

```python
df=data.drop(['Concentrations'], axis=1)
```

```python
df.head()
```

```python
from matplotlib import font_manager
sns.set()
fig,hm=plt.subplots(figsize=(10,6))



x=['A549 Cells (A)', 'A549 Cells (B)']

hm=sns.heatmap(data=df,yticklabels=x,annot=True,fmt=".2f",cmap='plasma',linewidth=1,linecolor='white',center=0,annot_kws={"size":10, "weight":"bold", 'family':'Times New Roman'}, square=True,cbar_kws=dict(use_gridspec=False,location="bottom")) 

plt.xticks(fontname="Times New Roman",fontsize=12,rotation=90,weight='bold')   
plt.yticks(fontname="Times New Roman",fontsize=12,rotation=0,weight='bold')
hm.figure.axes[1].tick_params(axis="y",labelsize=12,width=20)

cbar = hm.collections[0].colorbar

plt.setp(cbar.ax.get_yticklabels(), fontsize=12,rotation=0, fontfamily='Times New Roman', weight='bold')


hm1=hm.get_figure()
hm1.savefig('f:\BENZ-Toxicity H.png', bbox_inches='tight', dpi=300)
```
