```python
# importing the required libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Data set
df=pd.read_csv('D:\Data_h\ADME Data.csv')

df
```

```python
df.describe()
```

```python
df1=df.drop(['Ligand'], axis=1)
df1
```

```python
# Define the plot
fig, ax = plt.subplots(figsize=(5,10))

# Add title to the Heat map
title = "Pharma Heat Map"

# Set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl = ax.title
ttl.set_position([0.5,1.05])

# Hide ticks for X & Y axis
ax.set_xticks([])
ax.set_yticks([])

# Remove the axes
ax.axis('off')
# Use the heatmap function from the seaborn package
sns.heatmap(df1,annot= True,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax)

# Display the Pharma Heatmap
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
sns.heatmap(df1.corr(), cmap="YlGnBu", annot=True,  cbar=True,annot_kws={"size": 10, 'family':'Times New Roman'})
ax.set_xticklabels(ax.get_xticklabels(), fontname='Times New Roman')
ax.set_yticklabels(ax.get_xticklabels(), fontname='Times New Roman')
cbar = ax.collections[0].colorbar
plt.setp(cbar.ax.get_yticklabels(), fontsize=10, fontfamily='Times New Roman')
# new roman font
```

```python
plt.figure(figsize=(16,9))
 
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'fontfamily':'Times New Roman',
           'color':"k",
           'alpha':0.6, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
 
sns.heatmap(df1.corr(), annot = True, annot_kws= annot_kws)
```

```python
# Heatmap with keyword arguments (kwargs) parameter
 
plt.figure(figsize = (16,9))
 
kwargs = {'alpha':.9,'linewidth':10, 'linestyle':'--', 'rasterized':False, 'edgecolor':'w',  "capstyle":'projecting',}
 
ax = sns.heatmap(df1.corr(),annot=True,**kwargs)
```

```python
# Upper triangle heatmap
import numpy as np 
plt.figure(figsize=(16,9))
 
corr_mx = df1.corr() # correlation matrix
 
matrix = np.tril(corr_mx) # take lower correlation matrix
 
sns.heatmap(corr_mx,annot=True, mask=matrix)
```
