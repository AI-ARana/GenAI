```python
def O_iqr (data, feat):
    q25, q75=np.percentile(data[feat],q=[25,75])
    iqr=q75-q25
    u_lmt=q75+iqr*1.45
    l_lmt=q25-iqr*1.45
    return data[(data[feat]>u_lmt) | (data[feat]<l_lmt)].index.to_list()
    
```

```python
def d_dummy(data):
    return pd.get_dummies(data)
```

```python
X_cln=d_dummy(x)
```

```python
X_cln.info()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(X_cln.corr(),annot=True,cmap='RdYlGn')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_cln,y,test_size=0.2,random_state=1)
```

```python
from sklearn.matrics import r2_score
```

```python
from sklearn.tree import DecisionTreeRegressor

reg1 = DecisionTreeRegressor(max_depth=13)
reg1.fit(xtrain, ytrain)

# Prediction
ypred =  reg1.predict(xtest)

# Checking Accuracy
r2_score(ytest, ypred) 
```

```python
from sklearn.ensemble import RandomForestRegressor

reg2 = RandomForestRegressor(n_estimators=29, max_depth=17, random_state=1) # Criterion = Squared Error by default
reg2.fit(xtrain, ytrain)

# Predicting using Xtest Data
ypred = reg2.predict(xtest)

# Checking Accuracy
r2_score(ytest, ypred)
```
